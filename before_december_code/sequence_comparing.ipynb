```python
def read_fasta_file(path):
    '''
    used for load fasta data and transformd into numpy.array format
    '''
    fh=open(path)
    seq=[]
    for line in fh:
        if line.startswith('>'):
            continue
        else:
            seq.append(line.replace('\r\n',''))
    fh.close()
    matrix_data=np.array([list(e) for e in seq])
    return matrix_data
```

```python
def split_positive_negative(matrix_data):
    positive_data=matrix_data[xrange(len(matrix_data)/2)]
    negative_data=matrix_data[xrange(len(matrix_data)/2,len(matrix_data))]
    return positive_data,negative_data
```

```python
def cut_small_sequence(data_line,index,size):
    return "".join(data_line[index:index+size])
```

```python
def comparing_with_oneline(fragment,one_line,size):
    for index,data in enumerate(one_line):
        if fragment==cut_small_sequence(one_line,index,size):
            return False
        
    return True
```

```python
def comparing_line_with_subset(data_line,subset):
    save_result=[]
    for fragment in xrange(4,len(data_line)):
        for index,data in enumerate(data_line):
            if index<=len(data_line)-fragment:
                small_sequence=cut_small_sequence(data_line,index,fragment)
                num=np.sum([comparing_with_oneline(small_sequence,e,fragment) for e in subset])
                if num==len(subset):
                    save_result.append(small_sequence)
                else:
                    continue
    print save_result
    return save_result
```

```python
path='/home02/chenhuangrong/m6a_data.txt'
m6a_data=read_fasta_file(path)
positive_data,negative_data=split_positive_negative(m6a_data)
final_result=[]
for index,data in enumerate(positive_data):
    final_result.extend(comparing_line_with_subset(data,negative_data))
for index,data in enumerate(negative_data):
    final_result.extend(comparing_line_with_subset(data,positive_data))
print final_result
```
